```python
item = input("Enter item name: ")
price = float(input("Enter price: "))
quantity = float(input("Enter quantity: "))
subtotal = price * quantity
tax = subtotal * 0.05
total = tax + subtotal
print("-" * 15)
print(f'Item: {item}')
print(f'Subtotal: {subtotal}')
print(f'Tax (5%): {tax}')
print("-" * 15)
print(total)
```
