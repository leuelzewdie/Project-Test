```python
currency = input("What currency do you have in hand?Enter 'USD' or 'ETH': ").upper()
amount = float(input("Amount: "))
if currency == 'USD':
    amount = amount * 100
    print(f"Your amount is {amount}.")
elif currency == 'ETH':
    amount = amount * 0.01
    print(f"Your amount is {amount}.")
```
