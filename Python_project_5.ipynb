```python
grade = float(input("Enter your garde: "))
print("_" * 15)
if grade > 100:
    print("Invalid Score. Please enter 0-100")
elif grade == 100 or grade >= 90:
    print("Grade: A")
elif grade == 89 or grade >=80:
    print("Grade: B")
elif grade == 79 or grade >= 70:
    print("Grade: C")
elif grade == 69 or grade >=60:
    print("Grade: D")
elif grade == 0 or grade <=59:
    print("Grade: F")
else:
    print("Invalid Score. Please enter 0-100")
```
